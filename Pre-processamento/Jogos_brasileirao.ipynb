```python
import pandas as pd
```

# Importando dados

```python
df_jogos2013 = pd.read_csv('jogos_2013.csv', index_col=['id_arena','id_clube1',
                                                        'id_clube2','id_clube_vencedor',
                                                        'id_estado_vencedor','id_estado_clube1',
                                                        'id_estado_clube2'])
df_jogos2014 = pd.read_csv('jogos_2014.csv', index_col=['id_arena','id_clube1',
                                                        'id_clube2','id_clube_vencedor',
                                                        'id_estado_vencedor','id_estado_clube1',
                                                        'id_estado_clube2'])
df_jogos2015 = pd.read_csv('jogos_2015.csv', index_col=['id_arena','id_clube1',
                                                        'id_clube2','id_clube_vencedor',
                                                        'id_estado_vencedor','id_estado_clube1',
                                                        'id_estado_clube2'])
df_jogos2016 = pd.read_csv('jogos_2016.csv', index_col=['id_arena','id_clube1',
                                                        'id_clube2','id_clube_vencedor',
                                                        'id_estado_vencedor','id_estado_clube1',
                                                        'id_estado_clube2'])
df_jogos2017 = pd.read_csv('jogos_2017.csv', index_col=['id_arena','id_clube1',
                                                        'id_clube2','id_clube_vencedor',
                                                        'id_estado_vencedor','id_estado_clube1',
                                                        'id_estado_clube2'])
```

```python
df_arenas  = pd.read_csv('arenas.csv', index_col='id_arena')
df_clubes  = pd.read_csv('clubes.csv', index_col='id_clube')
df_estados = pd.read_csv('estados.csv', index_col='id_estado')
```

# Concatenar dados

```python
df_jogos2013.shape
```

```python
df_jogos = pd.concat([df_jogos2013, df_jogos2014, df_jogos2015, df_jogos2016, df_jogos2017], axis = 0)
```

```python
df_jogos.shape
```

```python
df_jogos.head()
```

```python
df_arenas.head()
```

```python
df_jogos = df_jogos.join(df_arenas, how='left')
```

```python
df_jogos = df_jogos.join(df_clubes, how='left', on=['id_clube1'], lsuffix='1', rsuffix='2')
df_jogos = df_jogos.join(df_clubes, how='left', on=['id_clube2'], lsuffix='1', rsuffix='2')
df_jogos.head()
```

```python
df_jogos = df_jogos.join(df_clubes, how='left', on=['id_clube_vencedor'])
df_jogos.head()
```

```python
df_jogos = df_jogos.join(df_estados, how='left', on=['id_estado_clube1'], lsuffix='1', rsuffix='2')
df_jogos = df_jogos.join(df_estados, how='left', on=['id_estado_clube2'], lsuffix='1', rsuffix='2')
df_jogos = df_jogos.join(df_estados, how='left', on=['id_estado_vencedor'])
df_jogos.head()
```

```python
df_jogos.rename(columns={"desc_arena":"arena", 
                         "desc_clube1":"clube1", 
                         "desc_clube2":"clube2", 
                         "desc_clube":"clube_vencedor", 
                         "desc_estado":"estado_vencedor", 
                         'sigla_estado':'sigla_estado_vencedor'}, inplace = True)
df_jogos.head()
```

```python
df_jogos.reset_index(drop=True, inplace=True)
df_jogos.head()
```

## Qual o clube que mais marcou gols em todas as temporadas?

```python
df_jogos.groupby(['clube1'])['qtd_gols_clube1'].sum().max()
```

```python
df_jogos.groupby(['clube2'])['qtd_gols_clube2'].sum().max()
```

```python
# O time do clube 1 marcou mais pontos então:
df_jogos.groupby(['clube1'])['qtd_gols_clube1'].sum().sort_values(ascending = False).head(1)
```

```python
if( df_jogos.groupby(['clube1'])['qtd_gols_clube1'].sum().max() > df_jogos.groupby(['clube2'])['qtd_gols_clube2'].sum().max() ):
  print(df_jogos.groupby(['clube1'])['qtd_gols_clube1'].sum().sort_values(ascending = False).head(1))
else:
  print(df_jogos.groupby(['clube2'])['qtd_gols_clube2'].sum().sort_values(ascending = False).head(1))
```

## Qual o clube que mais marcou gols na temporada de 2015?

```python
jogos2015 = df_jogos[df_jogos['data'].str.slice(start=6) == '2015']
if( jogos2015.groupby(['clube1'])['qtd_gols_clube1'].sum().max() > jogos2015.groupby(['clube2'])['qtd_gols_clube2'].sum().max() ):
  print(jogos2015.groupby(['clube1'])['qtd_gols_clube1'].sum().sort_values(ascending = False).head(1))
else:
  print(jogos2015.groupby(['clube2'])['qtd_gols_clube2'].sum().sort_values(ascending = False).head(1))
```

## Qual o jogo onde houveram mais gols, em todas as temporadas?

```python
df_jogos['qtd_total_gols'] = df_jogos['qtd_gols_clube1']+df_jogos['qtd_gols_clube2']
df_jogos.head()
```

```python
df_jogos.sort_values(by = 'qtd_total_gols', ascending = False).head(1)
```

## Qual foi o estado mais vitorioso (quantidade de partidas) em 2016?

```python
jogos2016 = df_jogos[df_jogos['data'].str.slice(start=6) == '2016']
jogos2016.groupby('estado_vencedor')['estado_vencedor'].count().sort_values(ascending=False)
```

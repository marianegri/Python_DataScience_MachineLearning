```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y_means = make_blobs(n_samples=300, centers=4, ## cluster_std: desvio padrão dos grupos 
                        cluster_std=0.60, random_state=0) ## random_state: semente do nº aleatorio
plt.scatter(X[:,0], X[:,1])
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=0).fit(X)
y_kmeans = kmeans.predict(X)

centers = kmeans.cluster_centers_

plt.scatter(X[:,0], X[:,1], c=y_kmeans )
plt.scatter(centers[:,0], centers[:,1], c ='black', alpha=0.5, s=200)
```
